```python
def convert_units():
    """
    Converts between different units of measurement.

    Supported conversions:
    - Length: meters <-> feet
    - Weight: kilograms <-> pounds
    - Volume: liters <-> gallons

    Returns:
        float: converted value
    """

    try:
        print("Choose conversion type:")
        print("1. Length")
        print("2. Weight")
        print("3. Volume")

        choice = int(input("Enter choice (1-3): "))

        value = float(input("Enter value to convert: "))

        if choice == 1:
            print("1. meters -> feet")
            print("2. feet -> meters")
            option = int(input("Choose option: "))

            if option == 1:
                result = value * 3.28084
                print("Result:", result, "feet")
            elif option == 2:
                result = value / 3.28084
                print("Result:", result, "meters")

        elif choice == 2:
            print("1. kg -> pounds")
            print("2. pounds -> kg")
            option = int(input("Choose option: "))

            if option == 1:
                result = value * 2.20462
                print("Result:", result, "pounds")
            elif option == 2:
                result = value / 2.20462
                print("Result:", result, "kg")

        elif choice == 3:
            print("1. liters -> gallons")
            print("2. gallons -> liters")
            option = int(input("Choose option: "))

            if option == 1:
                result = value * 0.264172
                print("Result:", result, "gallons")
            elif option == 2:
                result = value / 0.264172
                print("Result:", result, "liters")

        else:
            print("Invalid conversion type")

    except ValueError:
        print("Invalid input. Please enter numeric values.")
```

```python
convert_units()
```

```python
def find_sum(numbers):
    """
    Returns sum of numbers in a list.
    """
    return sum(numbers)

def find_average(numbers):
    """
    Returns average of numbers in a list.
    """
    return sum(numbers) / len(numbers)

def find_min(numbers):
    """
    Returns minimum number in list.
    """
    return min(numbers)

def find_max(numbers):
    """
    Returns maximum number in list.
    """
    return max(numbers)
try:
    operation = input("Choose operation (sum, avg, max, min): ")

    nums = list(map(float, input("Enter numbers separated by space: ").split()))

    if len(nums) == 0:
        raise ValueError("List is empty")

    if operation == "sum":
        print(find_sum(nums))

    elif operation == "avg":
        print(find_average(nums))

    elif operation == "max":
        print(find_max(nums))

    elif operation == "min":
        print(find_min(nums))

    else:
        print("Invalid operation")

except ValueError:
    print("Invalid input.")
```

Exercise on List Manipulation

```python
my_list = [1, 2, 3, 4, 5, 6]
def extract_every_other(lst):
    """
    Returns every other element from list.
    """
    return lst[::2]
extract_every_other(my_list)
```

```python
def get_sublist(lst, start, end):
    """
    Returns sublist from start to end (inclusive).
    """
    return lst[start:end+1]
get_sublist(my_list, 1, 4)
```

```python
def reverse_list(lst):
    """
    Returns reversed list using slicing.
    """
    return lst[::-1]
reverse_list(my_list)
```

```python
def remove_first_last(lst):
    """
    Removes first and last elements.
    """
    return lst[1:-1]
remove_first_last(my_list)
```

```python
def get_first_n(lst, n):
    """
    Returns first n elements.
    """
    return lst[:n]
get_first_n(my_list, 3)
```

```python
def get_last_n(lst, n):
    """
    Returns last n elements.
    """
    return lst[-n:]
get_last_n(my_list, 3)
```

```python
def reverse_skip(lst):
    """
    Returns elements starting from second-last
    in reverse order skipping one each time.
    """
    return lst[-2::-2]
reverse_skip([1,2,3,4,5,6])
```

Nested List Exercise 

```python
def flatten(lst):
    """
    Flattens one level of nested list.
    """
    result = []
    for sub in lst:
        for item in sub:
            result.append(item)
    return result
flatten([[1, 2], [3, 4], [5, 6]])
```

```python
def access_nested_element(lst, indices):
    """
    Access element using index path.
    """
    element = lst
    for i in indices:
        element = element[i]
    return element
lst = [[1,2,3],[4,5,6],[7,8,9]]
access_nested_element(lst,[1,2])
```

```python
def sum_nested(lst):
    """
    Returns sum of all numbers in nested list.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total
lst = [1, [2, 3], 4, 5]
sum_nested(lst)
```

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of elem in nested list.
    """
    result = []

    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        else:
            if item != elem:
                result.append(item)

    return result
lst = [1, [2, 3], 4, 5, 2]
remove_element(lst, 2)
```

```python
def find_max(lst):
    """
    Returns maximum value in nested list.
    """
    maximum = float("-inf")

    for item in lst:
        if isinstance(item, list):
            maximum = max(maximum, find_max(item))
        else:
            maximum = max(maximum, item)

    return maximum
lst = [1, [2, 3], 4, 5]
find_max(lst)
```

```python
def count_occurrences(lst, elem):
    """
    Counts occurrences of element in nested list.
    """
    count = 0

    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1

    return count
lst = [1, [2, 3], 4, 5, 2]
count_occurrences(lst, 2)
```

```python
def deep_flatten(lst):
    """
    Flattens nested list of any depth.
    """
    result = []

    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)

    return result
lst = [1, [2, [3, 4], 5], 6]
deep_flatten(lst)
```

```python
def average_nested(lst):
    """
    Returns average of nested list elements.
    """

    flat = deep_flatten(lst)

    return sum(flat) / len(flat)
lst = [1, [2, [3, 4], 5], 6]
average_nested(lst)
```

Experiment: How Fast is NumPy

Element-wise Addition

```python
import numpy as np
import time
```

```python
size = 1000000

list1 = list(range(size))
list2 = list(range(size))

start = time.time()

result = [list1[i] + list2[i] for i in range(size)]

end = time.time()

print("Python list addition time:", end-start)
```

```python
arr1 = np.arange(size)
arr2 = np.arange(size)

start = time.time()

result = arr1 + arr2

end = time.time()

print("NumPy addition time:", end-start)
```
